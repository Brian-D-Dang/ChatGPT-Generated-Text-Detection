```python
essays = []
essayIds = []
def retrieve_essays():
    for folder in folder_paths:
        # sets os path
        os.chdir(folder[0]) 
        # iterate through all files in directory ending with .txt
        for file in os.listdir(): 
            # Check whether file is in text format or not 
            if file.endswith(".txt"): 
                file_path = f"{folder[0]}\{file}"
                # call read text file function 
                essay_id = int(file.split('.')[0])
                with open(file_path, encoding='utf-8') as file:
                    data = file.read()
                essays.append(data)
                essayIds.append(essay_id)

retrieve_essays()
```

```python
essays_labels = []
for folder in folder_paths:
    # sets os path
    os.chdir(folder[0]) 
    # iterate through all files in directory ending with .txt
    for file in os.listdir(): 
        # Check whether file is in text format or not 
        if file.endswith(".txt"): 
            file_path = f"{folder[0]}\{file}"
            # call read text file function 
            essay_id = int(file.split('.')[0])
            label = folder[1]
            essays_labels.append({'Id': essay_id, 'Label': label})
essays_labels
```

```python
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(essays)
print(tfidf.get_feature_names_out())
```

```python
tfidf_df = pd.DataFrame(data=X.toarray(), columns=tfidf.get_feature_names_out())
labels = pd.DataFrame(essays_labels)
tfidf_df['Id'] = essayIds
tfidf_df = tfidf_df.join(labels.set_index('Id'), on="Id")
# tfidf_df['s']
```
